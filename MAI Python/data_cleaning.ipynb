# Import libraries

```python
import pandas as pd
import re
```

# Load Datasets

```python
# Load csv files
# Selected by Birds and loctions ('Victoria', 'Queensland', 'South Australia')
# Endangered-birds Dataset 
bird_df = pd.read_csv('./datasets/raw_Endangered-birds-records.csv', low_memory=False)
# LaTrobe endangered species list Dataset 
Labird_df = pd.read_csv('./datasets/raw_LaTrobe-endangered-birds.csv', low_memory=False)
```

# Check Differences among Datasets

```python
# Check column names
print(bird_df.columns)
print(Labird_df.columns)
# Check column differences
print(bird_df.columns.symmetric_difference(Labird_df.columns))
```

# Join Datasets

```python
combined_df = pd.concat([bird_df, Labird_df], ignore_index=True)
combined_df.head()
```

# Clean Dataset

```python
# 1. Drop all columns that all rows are empty
combined_df = combined_df.dropna(axis=1, how='all')

# Even though both datasets has the same columns, missing values for each column are different
# So removing columns must be done seperated at first 

# 2. Remove columns that mosly empty and not useful for Bird Detection
# Clean for reason (might be useful)
# If is not mention the rows are not meaningful
selected_df = combined_df.drop(['images','dcterms:modified', 'preparations',
                       'dataResourceUid', 'collectionCode', # ID is not useful
                       'occurrenceID', # This one is duplicate with Reference column
                       'dcterms:language', 'dcterms:license','rightsHolder',
                       'institutionID', 'datasetID', 'ownerInstitutionCode',
                       'dataGeneralizations', 'catalogNumber',
                       'recordedBy', 'individualCount','sex', # Sex, lifestage is mostly NaN
                       'lifeStage', 'occurrenceStatus', 'otherCatalogNumbers', # occurrenceStatus Same values
                       'occurrenceRemarks', 'associatedOccurrences', 'eventID', # Comments is not useful here
                       'fieldNumber', 'eventTime', 'startDayOfYear', 'year', 
                       'month','day', 'verbatimEventDate', # Object detection doesn't need Date
                       'eventDate', 'endDayOfYear',
                       'samplingProtocol', 'eventRemarks', 'locationID','continent', # Most rows empty
                       'country','countryCode', # Country(Code): Rows are all Australia
                       'municipality', 'locality', 'verbatimLocality', # Most rows are empty
                       'minimumElevationInMeters', 'maximumElevationInMeters',
                       'locationRemarks', 'verbatimElevation', # ^v Most rows are empty
                       'decimalLatitude', 'decimalLongitude', 'geodeticDatum','coordinateUncertaintyInMeters',
                       'coordinatePrecision', 'verbatimCoordinates', 'verbatimLatitude',
                       'verbatimLongitude', 'verbatimCoordinateSystem', 'verbatimSRS',
                       'georeferencedDate', 'georeferenceProtocol', 'georeferenceSources', # no usage of location
                       'identificationID', 'identifiedBy', 'dateIdentified', 
                       'identificationRemarks','taxonID', # No ID usage
                       'scientificNameID', 'acceptedNameUsage', 'nameAccordingTo', # Mostly empty
                       'higherClassification', 'specificEpithet','infraspecificEpithet', # Mostly empty
                       'taxonRank', 'scientificNameAuthorship', 
                       'nomenclaturalCode', 'taxonRemarks', 'recordedByID', 'identifiedByID', # Mostly empty
                       'provenance', 'sounds', 'recordID', 'rights', # Provenance, sounds, rights: Mostly empty
                       'images.1', 'dcterms:type', 'institutionCode', # Just ID not useful
                       'datasetName', 'raw_scientificName', # datasetName: Same value all rows, raw is duplicate
                      ], axis=1)

# 3. Preserved speicimen removed since we do object detection for the whole body of birds
# Not Eggs, feather, or bottled images
print(selected_df['basisOfRecord'].unique())
selected_df = selected_df[selected_df['basisOfRecord'] != 'PRESERVED_SPECIMEN']
# Then we drop it because we want all rows right now
selected_df = selected_df.drop(columns='basisOfRecord')

# 4. 'dynamicProperties' does not need anymore after removing 'BasicOfRecord', 
# it contains only ID or useless info. right now so delete it
# print(selected_df['dynamicProperties'].unique())
selected_df = selected_df.drop(columns='dynamicProperties')

# 5. Remove TaxonConceptID: URL links to scientific information 
# Which we do not need, common knowledge (name) is enough
selected_df = selected_df.drop(columns='taxonConceptID')

# 6. Remove seperated name columns
selected_df = selected_df.drop(columns=['kingdom','phylum','class','order','family','genus'])

# 7. Merge URL columns to "References" columns
# 7.1 recordNumber -> references
selected_df.loc[selected_df['recordNumber'].str.startswith('https', na=False), 'references'] = selected_df['recordNumber']
selected_df = selected_df.drop(columns='recordNumber')

# 7.2 associatedMedia -> references
selected_df.loc[selected_df['associatedMedia'].str.startswith('https', na=False), 'references'] = selected_df['associatedMedia']
selected_df = selected_df.drop(columns='associatedMedia')

# 7.3 identificationVerificationStatus -> references
# Extract url from a tag href using regexpr
# Captures URLs starting with http:// or https:// up until the next single quote (').
mask = selected_df['references'].isna()
extracted = selected_df.loc[mask, 'identificationVerificationStatus'].str.extract(r"href='(https?://[^']+)'")[0]
selected_df.loc[mask, 'references'] = extracted
selected_df = selected_df.drop(columns='identificationVerificationStatus')

# 7.4 Drop na for references (Image URL) after copy all URL to 'reference' column
selected_df = selected_df.dropna(subset=['references'])

# 8. Remove url 'references' duplicates
```
