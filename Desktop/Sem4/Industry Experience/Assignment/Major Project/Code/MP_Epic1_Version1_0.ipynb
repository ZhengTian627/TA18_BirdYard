Basic Cleanup

```python
def clean_text_column(plants, column_name, case='title', fillna='Unknown'):
    if case == 'title':
        plants[column_name] = plants[column_name].astype(str).str.strip().str.title()
    elif case == 'lower':
        plants[column_name] = plants[column_name].astype(str).str.strip().str.lower()
    elif case == 'upper':
        plants[column_name] = plants[column_name].astype(str).str.strip().str.upper()

    # replacing empty string with NaN
    plants[column_name] = plants[column_name].replace('', pd.NA)

    # Filling missing values
    if fillna is not None:
        plants[column_name] = plants[column_name].fillna(fillna)
```

```python
cols_to_title = [
    'Common Name', 'Plant Type', 'Water Needs', 'Climate Zones',
    'Light Needs', 'Soil Type', 'Soil Additional', 'Maintenance',
    'Height Ranges', 'Spread Ranges', 'Flower colour', 'Foliage Colour',
    'Butterfly Type', 'Image Location', 'Image Owner'
]

cols_to_lower = [
    'Perfume', 'Aromatic', 'Edible', 'Bird Attracting',
    'Greywater Tolerance', 'Native', 'Butterfly Attracting'
]

# Clean all title-case columns
for col in cols_to_title:
    clean_text_column(plants, col, case='title')

# Clean all lower-case columns
for col in cols_to_lower:
    clean_text_column(plants, col, case='lower')
```

Standarizing Binary Column

```python
def standardize_binary_column(plants, column_name):
    plants[column_name] = plants[column_name].str.lower().map({
        'yes': 'Yes',
        'no': 'No',
        'unknown': 'Unknown',
        'y': 'Yes',
        'n': 'No',
        '': 'Unknown',
        'nan': 'Unknown',
        pd.NA: 'Unknown'
    }).fillna('Unknown')

```

```python
#Applying it to binary columns
binary_cols = ['Perfume', 'Aromatic', 'Edible', 'Bird Attracting',
               'Greywater Tolerance', 'Native', 'Butterfly Attracting']

for col in binary_cols:
    standardize_binary_column(plants, col)
```

**Cleaning and Preparing Climate zones for Filtering**

```python
# Clean and split Climate Zones into a list
def clean_climate_zones(cell):
    if pd.isna(cell):
        return []
    return [zone.strip().title() for zone in str(cell).split(',')]

plants['Climate Zones Clean'] = plants['Climate Zones'].apply(clean_climate_zones)
```

```python
# Flatten all climate zone values into a set
all_zones = set(chain.from_iterable(plants['Climate Zones Clean']))
print("Unique Climate Zones:")
print(sorted(all_zones))
```

```python
#Removing Unknown Climates
plants = plants[plants['Climate Zones Clean'].apply(lambda x: 'Nan' not in x and x != [])]
```

Cleaning Soil Type

```python
def clean_soil_type(cell):
    if pd.isna(cell):
        return []
    return [soil.strip().title() for soil in str(cell).split(',')]

plants['Soil Type Clean'] = plants['Soil Type'].apply(clean_soil_type)
```

```python
unique_soils = sorted(set(chain.from_iterable(plants['Soil Type Clean'])))
print("Unique Soil Types:")
print(unique_soils)
```

```python
#removing Unknown
plants = plants[plants['Soil Type Clean'].apply(lambda x: 'Nan' not in x and x != [])]
```

Cleaning Water and Light Column

```python
# Reuse earlier cleaning function
def clean_text_column(plants, column_name, case='title', fillna='Unknown'):
    if case == 'title':
        plants[column_name] = plants[column_name].astype(str).str.strip().str.title()
    elif case == 'lower':
        plants[column_name] = plants[column_name].astype(str).str.strip().str.lower()
    elif case == 'upper':
        plants[column_name] = plants[column_name].astype(str).str.strip().str.upper()

    plants[column_name] = plants[column_name].replace('', pd.NA).fillna(fillna)

# Clean Water and Light columns
clean_text_column(plants, 'Water Needs')
clean_text_column(plants, 'Light Needs')
```

Recommendation based on Climate and soil

```python
climate_aliases = {
    'Cool Temperate': 'Temperate',
    'Warm Temperate': 'Temperate',
    'Dry Sub-Tropical': 'Dry Subtropical',
    'Dry Tropical': 'Dry Tropical',
    'Humid Sub-Tropical': 'Humid Subtropical',
    'Humid Tropical': 'Humid Tropical',
    'Arid': 'Desert'
}

def recommend_plants_by_climate_flexible(user_input):
    # Map input to cleaned value
    mapped_input = climate_aliases.get(user_input.title(), user_input.title())
    filtered_plants = plants[plants['Climate Zones Clean'].apply(lambda x: mapped_input in x)]
    return filtered_plants[['Botanical Name', 'Common Name', 'Climate Zones']]

# Try:
recommend_plants_by_climate_flexible('Cool Temperate').head(10)
```

```python
def recommend_plants_by_climate_and_soil(climate_zone, soil_type):
    climate_zone = climate_zone.strip().title()
    soil_type = soil_type.strip().title()

    filtered_plants = plants[
        plants['Climate Zones Clean'].apply(lambda x: climate_zone in x) &
        plants['Soil Type Clean'].apply(lambda x: soil_type in x)
    ]

    return filtered_plants[['Botanical Name', 'Common Name', 'Climate Zones', 'Soil Type']]
```

```python
import zipfile
from pathlib import Path

# Prepare Streamlit app script
streamlit_code = '''
import pandas as pd
import streamlit as st
from ast import literal_eval

st.set_page_config(page_title="Plant Recommender", layout="wide")
st.title("ðŸŒ¿ Waterwise Plant Recommendation System")

@st.cache_data
def load_data():
    plants = pd.read_csv("cleaned_waterwise_plants.csv")
    plants['Climate Zones Clean'] = plants['Climate Zones Clean'].apply(literal_eval)
    plants['Soil Type Clean'] = plants['Soil Type Clean'].apply(literal_eval)
    return plants

plants = load_data()

# Dropdowns
climate = st.selectbox("Select Climate Zone", sorted(set([z for sub in plants['Climate Zones Clean'] for z in sub])))
soil = st.selectbox("Select Soil Type", sorted(set([z for sub in plants['Soil Type Clean'] for z in sub])))

# Filter function
def recommend_plants_with_info(climate_zone, soil_type):
    filtered_plants = plants[
        plants['Climate Zones Clean'].apply(lambda x: climate_zone in x) &
        plants['Soil Type Clean'].apply(lambda x: soil_type in x)
    ]

    return filtered_plants[[
        'Botanical Name', 'Common Name', 'Climate Zones', 'Soil Type',
        'Water Needs', 'Light Needs', 'Height Ranges', 'Spread Ranges',
        'Maintenance', 'Flower colour', 'Edible', 'Native', 'Bird Attracting'
    ]]

# Filter and display
results = recommend_plants_with_info(climate, soil)
st.subheader(f"Recommended Plants for {climate} Climate and {soil} Soil")
if results.empty:
    st.warning("No plants found for this combination.")
else:
    st.dataframe(results, use_container_width=True)
'''

# Save files
Path("streamlit_app").mkdir(parents=True, exist_ok=True)
with open("streamlit_app/app.py", "w") as f:
    f.write(streamlit_code)

# Copy CSV to the app folder
plants.to_csv("streamlit_app/cleaned_waterwise_plants.csv", index=False)

# Zip the folder
zip_path = "plant_recommender_streamlit_app.zip"
with zipfile.ZipFile(zip_path, "w") as zipf:
    for file in Path("streamlit_app").glob("*"):
        zipf.write(file, arcname=file.name)

zip_path
```
